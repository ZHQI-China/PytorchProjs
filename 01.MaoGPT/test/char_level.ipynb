```python
with open('../data/Mao1.txt','r',encoding='utf-8') as f:
    text = f.read()
len(text),text[:100]
```

```python
zh_chars=sorted(set(text))
vocab_size = len(zh_chars)
vocab_size,zh_chars[:100]
```

```python
stoi = {ch:i for i,ch in enumerate(zh_chars)}
itos = {i:ch for i,ch in enumerate(zh_chars)}

encode = lambda s:[stoi[c] for c in s]
decode = lambda l:''.join([itos[i] for i in l])

encode('你好，模型'),decode([210, 703, 2823, 79, 1771])
# 中文tokenizer: https://blog.csdn.net/qq_43426908/article/details/134748200
```

```python
import torch
data = torch.tensor(encode(text),dtype=torch.long)
print(data.shape)
print(data[:100])
```

```python
n = int(0.9*len(data))
train_data = data[:n]
val_data = data[n+1:]
```
